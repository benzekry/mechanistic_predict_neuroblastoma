```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
from stat_tools_python.cox_regression import cox_regression
from stat_tools_python import survival_analysis
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# Load data

```python
df = pd.read_excel("../data/2019_08_27_database.xls", index_col=0)
df = df.drop(23) # patient 22 removed because no data on PT dimensions
if  df['sexe'].str.contains("fille").sum() >0:
    df['sexe'] = (df['sexe']=="fille")*1
df = df.rename(columns={"age(mois)": "age"})
df = df.rename(columns={"sexe": "sex"})
df = df.loc[df.index.dropna()] # drop last 2 rows that were used in excel file to compute median
```

Remove patients with positive MIBG but no SIOPEN

```python
idx_MIBG_1 = (df['MIBG_neg'] == 1)
print(sum(idx_MIBG_1))
df         = df.loc[~idx_MIBG_1, :]
```

# Kaplan-Meier analysis

```python
output_dir = "statistical_analysis"
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
```

## OS and PFS

```python
km = KaplanMeierFitter()
km.fit(df['OS'], df['deces'])
print("Median OS = %.3g months" % km.median_survival_time_)
print("OS à 3 ans = %.3g%%" % (km.predict(3*12)*100))
print("OS à 5 ans = %.3g%%" % (km.predict(5*12)*100))
ax = km.plot(show_censors=True, ci_show=False, label="Overall survival")
km.fit(df['PFS'], df['progression'])
print("Median PFS = %.3g months" % km.median_survival_time_)
print("PFS à 3 ans = %.3g%%" % (km.predict(3*12)*100))
print("PFS à 5 ans = %.3g%%" % (km.predict(5*12)*100))
km.plot(ax=ax, show_censors=True, ci_show=False, label="Progression-free survival")
plt.ylim(0,1)
ax.set_yticklabels([0, 20, 40, 60, 80, 100]);
plt.ylabel("Survival (%)");
plt.xlabel("Months")
legend = ax.get_legend()
plt.setp(legend.get_texts(), fontsize=13);
legend.get_frame().set_linewidth(0.0)
plt.plot([3*12, 3*12], [0, 1], '--k');
plt.plot([5*12, 5*12], [0, 1], '--k');
plt.savefig(os.path.join(output_dir,"OS_PFS.pdf"), bbox_inches='tight')
```

## Group analysis PFS

### Functions for logrank testing

```python
def log_rank_pfs_dicho(df, variable, labels=["0", "1"], filename="pfs_dicho.pdf"):
    km.fit(df.loc[df[variable]==1, 'PFS'], df.loc[df[variable]==1, 'progression'], label=labels[1])
    ax = km.plot(ci_show=False, show_censors=True)
    km.fit(df.loc[df[variable]==0, 'PFS'], df.loc[df[variable]==0, 'progression'], label=labels[0])
    km.plot(ax=ax, ci_show=False, show_censors=True)
    plt.ylim(0,1)
    ax.set_yticklabels([0, 20, 40, 60, 80, 100]);
    plt.ylabel("PFS");
    plt.xlabel("Months")
    legend = ax.get_legend()
    plt.setp(legend.get_texts(), fontsize=18);
    legend.get_frame().set_linewidth(0.0)
    results = logrank_test(
        df.loc[df[variable]==1, 'PFS'], 
        df.loc[df[variable]==0, 'PFS'],
        df.loc[df[variable]==1, 'progression'],
        df.loc[df[variable]==0, 'progression']
        )
#     print('Logrank test, p = %.3g' % results.p_value)
    plt.text(0.05, 0.1, s="p = {:.3g}".format(results.p_value), transform=ax.transAxes, fontsize=16)
    if not os.path.exists(os.path.join(output_dir, "PFS")):
        os.mkdir(os.path.join(output_dir, "PFS"))
    plt.savefig(os.path.join(output_dir, "PFS", filename+".pdf"), bbox_inches='tight')
```

```python
log_rank_pfs_dicho(df, 'sex', labels=["Boys", "Girls"], filename="sex")
```

### Tumor size

```python
for idx in df.index:
    diam_1     = df.loc[idx, 'T1(mm)']
    diam_2     = df.loc[idx, 'T2(mm)']
    diam_3     = df.loc[idx, 'T3(mm)']
    if np.isnan(diam_3): # if only two diameters, assume the smallest of the first two as third value
        diam_3 = min(diam_1, diam_2)
    S_diag_mm3 = 4/3*np.pi*diam_1/2*diam_2/2*diam_3/2 # (mm3)
    df.loc[idx, 'S_diag_cm3'] = S_diag_mm3/1000
```

## Group analysis OS

### Functions for logrank testing

```python
def log_rank_os_dicho(df, variable, labels=["0", "1"], filename="survival"):
    km.fit(df.loc[df[variable]==1, 'OS'], df.loc[df[variable]==1, 'progression'], label=labels[1])
    ax = km.plot(ci_show=False, show_censors=True)
    km.fit(df.loc[df[variable]==0, 'OS'], df.loc[df[variable]==0, 'progression'], label=labels[0])
    km.plot(ax=ax, ci_show=False, show_censors=True)
    # Cosmetics
    plt.ylim(0,1)
    ax.set_yticklabels([0, 20, 40, 60, 80, 100]);
    plt.ylabel("OS");
    plt.xlabel("Months")
    legend = ax.get_legend()
    plt.setp(legend.get_texts(), fontsize=18);
    legend.get_frame().set_linewidth(0.0)
    results = logrank_test(
        df.loc[df[variable]==1, 'OS'], 
        df.loc[df[variable]==0, 'OS'],
        df.loc[df[variable]==1, 'deces'],
        df.loc[df[variable]==0, 'deces']
        )
#     print('Logrank test, p = %.3g' % results.p_value)
    plt.text(0.05, 0.1, s="p = {:.3g}".format(results.p_value), transform=ax.transAxes, fontsize=16)
    if not os.path.exists(os.path.join(output_dir, "OS")):
        os.makedirs(os.path.join(output_dir, "OS"))
    plt.savefig(os.path.join(output_dir, "OS", filename+".pdf"), bbox_inches='tight')
```
